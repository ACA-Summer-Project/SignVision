```python
"""1: Lists and Loops
Write a Python program that performs the following:"""
#1.1 Define the list: numbers = [15, 8, 22, 7, 31, 4, 17]
numbers = [15, 8, 22, 7, 31, 4, 17]
#1.2 Print all even numbers from the list
for x in numbers:
  if x%2==0:
    print(x)
#1.3 Create a new list containing the square of each odd number
new_list=[]
for y in numbers:
  if y%2!=0:
    new_list.append(y*y)
#1.4 Print the new list
print(new_list)
```

```python
"""2: Strings and Dictionaries
Write a program to count how many times each word appears in a sentence."""
#2.1 Given the sentence: "The book was interesting because the book covered many topics and the topics discussed in the book were engaging"
sentence = "The book was interesting because the book covered many topics and the topics discussed in the book were engaging"
#2.2 Convert the sentence to lowercase and split it into words.
words= sentence.lower().split()
#2.3 Create a dictionary where: o Keys are unique words o Values are the word counts
Dict={}
for x in words:
  if x in Dict:
    Dict[x]+=1
  else:
    Dict[x]=1
#2.4 Print the dictionary.
print(Dict)
```

```python
"""3: Functions and Conditional Logic
Define a function called is_prime(n) that checks if a number is prime."""
def is_prime(n):
  if n<2:
    return 0
  for i in range(2, int(n**0.5)+1):
    if n%i==0:
      return 0
    return 1
#3.1 Given the list: nums = [2, 4, 5, 10, 13, 17, 20, 23]
nums = [2, 4, 5, 10, 13, 17, 20, 23]
#3.2 Use the function to check each number in the list. Store and print only the prime numbers in a new list.
list_prime= []
for x in nums:
  if is_prime(x):
    list_prime.append(x)
print(list_prime)
```

```python
"""4: NumPy Array Operations and Statistics
Use NumPy to simulate and analyze student test scores."""
import numpy as np
#4.1 Generate a 2D NumPy array scores of shape (10, 5) representing scores of 10 students in 5 subjects, with random integers between 50 and 100.
arr= np.random.randint(50, 101, size=(10, 5))
print(arr)
#4.2.1 Compute the average score per student.
average= np.mean(arr, axis=1)
print(average)
#4.2.2 Compute the maximum score per subject.
max_score= np.max(arr, axis=0)
print(max_score)
#4.2.3 Add a new column with each student's average score.
average_column= np.array(average).reshape(-1,1)
scores= np.hstack((arr, average_column))
#4.2.4 Print the final array.
print(scores)
```

```python
 #9: Visualizing Student Scores
 import matplotlib.pyplot as plt
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [85, 78, 92, 88, 76],
    'Science': [90, 82, 89, 94, 75],
    'English': [78, 85, 88, 80, 82],
    'History': [84, 80, 91, 79, 77]
}
df=pd.DataFrame(data)
#9.1. Create a bar chart showing each student’s Math score
plt.bar(df['Name'], df['Math'])
plt.title('Math Scores by Student')
plt.xlabel('Student')
plt.ylabel('Math Score')
plt.show()
#2. Create a line plot comparing all 4 subjects for each student (one line per subject).
subjects = ['Math', 'Science', 'English', 'History']
for subject in subjects:
    plt.plot(df['Name'], df[subject], marker='o', label=subject)
plt.title('Student Performance Across Subjects')
plt.xlabel('Student')
plt.ylabel('Score')
plt.legend(title='Subjects')
plt.show()
```

```python
#10: Visualizing Random Data Distribution
#10.1 Generate 1000 random numbers from a normal distribution with mean = 50 and std = 10.
mean = 50
sd = 10
data = np.random.normal(mean, sd, 1000)
#10.2. Plot a histogram of the data
plt.hist(data, bins=20, edgecolor='black')
plt.axvline(x=np.mean(data),color='red',label='Mean')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
