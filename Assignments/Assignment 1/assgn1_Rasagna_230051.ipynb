```python
def isprime(n):
    count=0
    for i in range(1,n):
        if(n%i==0):
            count+=1
    return not count>1
nums = [2, 4, 5, 10, 13, 17, 20, 23]
newlist=[]
for x, x in enumerate(nums):
    if isprime(x):
        newlist.append(x)
print(newlist)
```
