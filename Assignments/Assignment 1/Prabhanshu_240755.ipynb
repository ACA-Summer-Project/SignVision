# Question 3

<h2>Defining a function called is_prime(x) that checks if a number is prime.</h2>

```python
def is_prime(x):
    check = 1
    for i in range(2, x):
        if x % i == 0:
            check = 0
            break
    if check == 1:
        return True
    else:
        return False
```

<h2>3.1 Defining the list</h2>

```python

nums = [2, 4, 5, 10, 13, 17, 20, 23] 
```

<h2>3.2 Using the function to check each number in the list.</h2>

```python
primes = [num for num in nums if is_prime(num)==True]
```

<h2>3.3 Print only the prime numbers in a new list.</h2>

```python
print(f"The list of primes in the given list is : {primes}")
```

<h2>6.2 Adding TotalValue column to the DataFrame</h2>

```python
a = np.array(arr)
Total_ = df['Price'] * df['Quantity']
Total = np.array(Total_)
Total_Value = Total.reshape(8,1)
new = np.append(a , Total_Value , axis=1)
df = pd.DataFrame(new , columns = ['Product' , 'Price' , 'Quantity', 'Total Value'])
df
```

<h2>6.3 Sorting the given DataFrame by TotalValue</h2>

```python
df = df.sort_values(by = 'Total Value', ascending = False)
df
```

```python
df['Price'] = pd.to_numeric(df['Price'])
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Range'] = pd.cut(df['Price'] , bins= [0,2000,5000,10000,100000])
df
```

```python
average_value = df.groupby('Range')['Quantity'].mean()
print(average_value)
```
