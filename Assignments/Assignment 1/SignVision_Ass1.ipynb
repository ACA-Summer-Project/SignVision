```python
import math
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True
```

```python
nums = [2, 4, 5, 10, 13, 17, 20, 23]
prime_nums = [num for num in nums if is_prime(num)]
print("Prime numbers:", prime_nums)
```

```python
import numpy as np
scores = np.random.randint(50, 101, size=(10, 5))
print("Scores array:\n", scores)
```

```python
student_averages = np.mean(scores, axis=1)
print("\nAverage Score Per Student:\n", student_averages)
subject_max = np.max(scores, axis=0)
print("\nMaximum Score Per Subject:\n", subject_max)
final_scores = np.column_stack((scores, student_averages))
print("\nFinal Scores with Student Averages:\n", final_scores)
```

```python
import csv
header = ["Subject1", "Subject2", "Subject3", "Subject4", "Subject5", "Average"]

filename = "student_scores.csv"

with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    for row in final_scores:
        writer.writerow(np.round(row, 2)) 

print(f"\nData saved to {filename}")
```

```python
column_names = [
    'Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG',
    'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'NumVesselsFluoro',
    'Thalassemia', 'HeartDisease'
]

file = "/home/priyanshu/processed.cleveland.data"
df = pd.read_csv(file, header=None, names=column_names)

print(df.head())
```

```python
df.replace('?', np.nan, inplace=True)

cols_to_convert = ['Age', 'Sex', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 
                   'Oldpeak', 'NumVesselsFluoro', 'Thalassemia', 'HeartDisease']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df.info())

```

```python
print("\nMissing values per column:\n", df.isnull().sum())
```

```python
mode_vessels = df['NumVesselsFluoro'].mode()[0]
df['NumVesselsFluoro'] = df['NumVesselsFluoro'].fillna(mode_vessels)

mode_thal = df['Thalassemia'].mode()[0]
df['Thalassemia'] = df['Thalassemia'].fillna(mode_thal)

print("Missing values after imputation:")
print(df[['NumVesselsFluoro', 'Thalassemia']].isnull().sum())
```

```python
total_cols = df.shape[1]

threshold = total_cols - 2

df_cleaned = df.dropna(thresh=threshold)

print(f"Rows before dropping: {len(df)}")
print(f"Rows after dropping: {len(df_cleaned)}")
```

```python
avg_cholesterol = df.groupby('HeartDisease')['Cholesterol'].mean()
print("Average Cholesterol level by HeartDisease status:\n", avg_cholesterol)
```

```python
import matplotlib.pyplot as plt

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [85, 78, 92, 88, 76],
    'Science': [90, 82, 89, 94, 75],
    'English': [78, 85, 88, 80, 82],
    'History': [84, 80, 91, 79, 77]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Math'], color='skyblue')
plt.title('Math Scores of Students')
plt.xlabel('Student Name')
plt.ylabel('Math Score')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
for subject in df.columns:
    plt.plot(df.index, df[subject], marker='o', label=subject)

plt.title('Student Performance Across Subjects')
plt.xlabel('Student')
plt.ylabel('Score')
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Subject')
plt.show()
```

```python
mean = 50
std_dev = 10
size = 1000

random_numbers = np.random.normal(loc=mean, scale=std_dev, size=size)
```

```python
plt.figure(figsize=(8,5))
plt.hist(random_numbers, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data (mean=50, std=10)', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(random_numbers, bins=20, color='skyblue', edgecolor='black')

plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean}')

plt.title('Histogram of Normally Distributed Data', fontsize=14, fontweight='bold')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.legend()

plt.show()
```
