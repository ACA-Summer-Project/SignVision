```python
def is_prime(n):
    if n<2:
        return 0
    for i in range(2,int(n/2+1)):
        if n%i==0:
            return 0
    return 1
numbers = [2,4,5,10,13,17,20,23]
prime_numbers = []
for i in numbers:
    if is_prime(i):
        prime_numbers.append(i)
print(prime_numbers)
```

```python
import numpy as np
scores = np.random.randint(50, 101, size=(10, 5))
avg = np.mean(scores, axis=1)
max = np.max(scores, axis=0)
final_array = np.hstack((scores, avg.reshape(-1, 1)))
print("Scores:\n", final_array)
```

```python
import pandas as pd
cols = ['Subject1','Subject2','Subject3','Subject4','Subject5','Average']
df = pd.DataFrame(final_array,columns = cols)
df.to_csv('scores.csv', index = False)
```

```python
import matplotlib.pyplot as plt
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [85, 78, 92, 88, 76],
    'Science': [90, 82, 89, 94, 75],
    'English': [78, 85, 88, 80, 82],
    'History': [84, 80, 91, 79, 77]
}
df = pd.DataFrame(data)
plt.bar(df['Name'], df['Math'])
plt.title('Math Scores by Student')
plt.xlabel('Student')
plt.ylabel('Math Score')
plt.show()
subjects = ['Math', 'Science', 'English', 'History']
for subject in subjects:
    plt.plot(df['Name'], df[subject], marker='o', label=subject)
plt.title('Student Performance')
plt.xlabel('Student')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
mean = 50
std = 10
data = np.random.normal(mean, std, size=1000)
mean_value = np.mean(data)
plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.axvline(mean_value, color='red', label = 'mean')
plt.title("Histogram of Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
plt.show()
```
