```python
import numpy as np
import matplotlib.pyplot as plt
```

# 10.1 Generating 1000 random numbers from a normal distribution with mean = 50 and std = 10.

```python
data = np.random.normal(loc = 50 , scale = 10 , size=1000)
x=data.mean()
```

# 10.2 Plotting a histogram of the data.
# 10.3 Customizing the plot as directed :

```python
plt.hist(data, edgecolor = 'Black' , bins = 20)
plt.axvline(x , color = 'Yellow' , linestyle = 'dashed')
plt.xticks(np.linspace(0,80,10))
plt.yticks(np.linspace(0,140,10))
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
