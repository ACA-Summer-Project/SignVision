```python
plt.plot(df['Name'] , df['Math'] , label = 'Math')
plt.plot(df['Name'] , df['Science'] , label = 'Science')
plt.plot(df['Name'] , df['English'] , label = 'English')
plt.plot(df['Name'] , df['History'] , label = 'History')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.legend()
plt.show()
```
