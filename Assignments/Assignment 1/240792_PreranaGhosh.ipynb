```python
def is_prime(n):
    result = False
    if n > 1:
        for i in range(2, n):
            if i%2 == 0:
                return result
        else:
            return True
    return result

list = [2, 4, 5, 10, 13, 17, 20, 23]
primes = []

for x in list:
    if is_prime(x):
        primes.append(x)

print(primes)
```

```python
import pandas as pd

arr = {'Product': ['Soap', 'Mobile', 'TV', 'Shoes', 'Books', 'Pencils', 'Car', 'Dress'],
        'Price': [120.00, 17000.00, 120000.00, 500.00, 450.00, 50.00, 900000.00, 745.00],
        'Quantity': [10, 2, 3, 14, 20, 100, 5, 34]}

df = pd.DataFrame(arr)

df['TotalValue'] = df['Price']*df['Quantity']

new_df = df.sort_values(by = 'TotalValue', ascending = True)

avg_quantity_per_price = df.groupby('Price')['Quantity'].mean()

print(new_df)

print(avg_quantity_per_price)
```

```python
import numpy as np
import csv

arr_scores = np.random.randint(50, 101, size = (10, 5))

avg_score_per_student = np.mean(arr_scores, axis = 1).reshape(-1, 1)

arr_final = np.hstack((arr_scores, avg_score_per_student))

header = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Average']

with open('StudentFinalScores.csv', mode = 'w') as file:
    newfile = csv.writer(file)
    newfile.writerow(header)
    newfile.writerows(arr_final)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dict = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Math': [85, 78, 92, 88, 76],
        'Science': [90, 82, 89, 94, 75],
        'English': [78, 85, 88, 80, 82],
        'History': [84, 80, 91, 79, 77]}

df = pd.DataFrame(dict)

plt.figure(figsize = (4, 5))
plt.bar(df['Name'], df['Math'])
plt.title('Math Scores')
plt.ylabel('Math Score')
plt.xlabel('Student Name')
plt.show()

plt.figure(figsize = (6, 5))

sub_arr = ['Math', 'Science', 'English', 'History']
for sub in sub_arr:
    plt.plot(df['Name'], df[sub], label = sub)
    plt.title('Overall performance comparison')
    plt.ylabel('Total Score')
    plt.xlabel('Student Name')
    plt.legend()
    plt.show()
```

```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

import pandas as pd
import numpy as np

#fetch dataset
heart_disease = fetch_ucirepo(id = 45)

#data(as pandas dataframes)
X = heart_disease.data.features
Y = heart_disease.data.targets

#metadata
print(heart_disease.metadata)

#variable information
print(heart_disease.variables)

info = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'NumVesselsFluoro', 'Thalassemia', 'HeartDisease']
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
df = pd.read_csv(url, names = info)

df.replace('?', np.nan, inplace = True)

for col in ['RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'NumVesselsFluoro', 'Thalassemia']:
    df[col] = pd.to_numeric(df[col], errors = 'coerce')

print(df.isnull().sum())

df['NumVesselsFluoro'].fillna(df['NumVesselsFluoro'].median(), inplace = True)
df['Thalassemia'].fillna(df['Thalassemia'].mode()[0], inplace = True)

df = df[df.isnull().sum(axis = 1) <= 2]

df['HeartDisease'] = df['HeartDisease'].astype(int)
df['HeartDisease'] = df['HeartDisease'].apply(lambda a: 1 if a > 0 else 0)

HeartDisease = df.groupby('HeartDisease')['Cholesterol'].mean()
print(HeartDisease)



                            
                            
    
```
