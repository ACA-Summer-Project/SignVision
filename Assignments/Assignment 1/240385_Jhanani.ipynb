```python
#Question2

#Defining sentence
sentence="The book was interesting because the book covered many topics and the topics discussed in the book were engaging"
#Converting sentence to lower words
sentence =sentence.lower()
print(sentence)
#Splitting sentence into words
words=sentence.split(" ")
new_words = set(words)

#Creating empty dictionary
words_count_map = {}

for i in new_words:
    words_count_map[i]=0
    for j in words:
        if i==j:
            words_count_map[i]+=1

#Printing dictionary
print(words_count_map)
```

```python
#Question3

#Defining function to check prime numbers
def is_prime(n):
    cnt=0
    for i in range(1,n//2+1):
        if not n%i:
            cnt+=1
        if cnt>1:
            break
    if cnt==1:
        return 1
    else:
        return 0

nums=[2,4,5,10,13,17,20,23]

#Creating list of prime numbers
new_list=[]

for i in nums:
    if is_prime(i):
        new_list.append(i)

print("Prime numbers are :", new_list)
```

```python
#Question5

import numpy as np

#Creating array of numbers of given dimension
array=np.random.randint(low=50,high=100,size=(10,5))

#Computing averages
averages=np.mean(array,axis=1)

#Computing maximum per subject
max=np.max(array,axis=0)

#Creating new column of averages
array=np.column_stack((array,averages))
data = array.tolist()

import csv
header = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Average']
file_path = 'student_scores.csv'
print(data)
# Open the file in write mode
with open(file_path, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header
    writer.writerow(header)

    # Write the data
    writer.writerows(data)
```

```python
#Question6

import pandas as pd

data = {
    'Product': ['Nike', 'Apple', 'Raymond','Crocs','CocaCola','Lenovo','Duracell','Dell'],
    'Price':[300.1,220.5,319.1,385.2,564.2,470.5,421.3,891.2],
    'Quantity': [1,4,5,6,7,2,9,3]
}

df = pd.DataFrame(data)
df['TotalValue']=df['Price']*df['Quantity']
print(df)
df=df.sort_values(by='TotalValue',ascending=False)
print(df)
df= df.groupby('Price')['Quantity'].mean().reset_index()
print(df)
```

```python
#Question8

import pandas as pd
import numpy as np

df = pd.read_csv('processed.cleveland.data')
df.columns=['Age','Sex','ChestPainType','RestingBP','Cholesterol','FastingBS','RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope','NumVesselsFluoro','Thalassemia','HeartDisease']
# Replace '?' with NaN
df = df.replace('?', np.nan)
print(df)
# Convert all columns to numeric
df = df.apply(pd.to_numeric)
print(df)

missing_counts = df.isna().sum()
print(missing_counts)

df['NumVesselsFluoro'] = df['NumVesselsFluoro'].fillna(df['NumVesselsFluoro'].median())
df['Thalassemia'] = df['Thalassemia'].fillna(df['Thalassemia'].median())
print(df)
df = df[df.isna().sum(axis=1) <= 2]
avg_cholesterol = df.groupby('HeartDisease')['Cholesterol'].mean()

print(avg_cholesterol)
```

```python
#Question9
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
data = [
    ["Alice", 85, 90, 78, 84],
    ["Bob", 78, 82, 85, 80],
    ["Charlie", 92, 89, 88, 91],
    ["David", 88, 94, 80, 79],
    ["Eva", 76, 75, 82, 77]
]

df = pd.DataFrame(data, columns = ["Name", "Math", "Science", "English", "History"])
print(df)

labels = [row[0] for row in data]
values = [row[1] for row in data]

# Plot the bar graph
df.set_index('Name')['Math'].plot(kind='bar', title='Math')

plt.ylabel('Math Marks')
plt.xlabel('Students')
plt.tight_layout()
plt.show()

#Plot line plot
df.set_index('Name').plot(kind='line', marker='o', title='Student Scores Across Subjects')

# Add labels and formatting
plt.ylabel('Marks')
plt.xlabel('Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
