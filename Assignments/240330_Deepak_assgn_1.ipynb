### Qusetion 2

```python
sentence = "The book was interesting because the book covered many topics and the topics discussed in the book were engaging"
words = sentence.lower().split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print(word_count)
```

### Question 3

```python

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n-1):
        if n % i == 0:
            return False
    return True

nums = [2, 4, 5, 10, 13, 17, 20, 23]
prime_nums = [num for num in nums if is_prime(num)]

print(prime_nums)
```

### Question 4

```python
import numpy as np

scores = np.random.randint(50, 101, size=(10, 5))
print("Original Scores (10 students, 5 subjects):\n", scores)

average_per_student = np.mean(scores, axis=1)
print("\nAverage score per student:\n", average_per_student)

max_per_subject = np.max(scores, axis=0)
print("\nMaximum score per subject:\n", max_per_subject)

final_array = np.column_stack((scores, average_per_student))
print("\nFinal array with average scores as a new column:\n", final_array)
```

### Question 6

```python
import pandas as pd

data = {
    "Product": ["Apple", "Banana", "Milk", "Bread", "Eggs", "Butter", "Cheese", "Juice"],
    "Price": [1.2, 0.5, 1.5, 2.0, 0.2, 2.5, 3.0, 1.8],
    "Quantity": [10, 20, 5, 7, 50, 3, 4, 6]
}

df = pd.DataFrame(data)

df["TotalValue"] = df["Price"] * df["Quantity"]
sorted_df = df.sort_values(by="TotalValue", ascending=False)

avg_quantity_by_price = df.groupby("Price")["Quantity"].mean()

print("Original DataFrame with TotalValue:\n", df, "\n")
print("Sorted by TotalValue (descending):\n", sorted_df, "\n")
print("Average Quantity grouped by Price:\n", avg_quantity_by_price)
```

### Question 9

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Math": [85, 78, 92, 88, 76],
    "Science": [90, 82, 89, 94, 75],
    "English": [78, 85, 88, 80, 82],
    "History": [84, 80, 91, 79, 77]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.bar(df["Name"], df["Math"], color='skyblue')
plt.title("Math Scores of Students")
plt.xlabel("Student Name")
plt.ylabel("Math Score")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
plt.figure(figsize=(10, 6))
subjects = ["Math", "Science", "English", "History"]

for i in range(len(df)):
    plt.plot(subjects, df.iloc[i, 1:], marker='o', label=df.iloc[i, 0])

plt.title("Student Performance in All Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.legend(title="Students")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Question 10

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=10, size=1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')

mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean_value:.2f}")

plt.title("Histogram of Normally Distributed Data (mean=50, std=10)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
