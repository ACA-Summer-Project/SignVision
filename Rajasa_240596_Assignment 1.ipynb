```python
 line = "The book was interesting because the book covered many topics and the topics discussed in the book were engaging"
```

```python
words = line.lower().split()
```

```python
word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
```

# Question 3

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
nums = [2, 4, 5, 10, 13, 17, 20, 23]
```

```python
prime_nums = [num for num in nums if is_prime(num)]
```

```python
print("Prime numbers:", prime_nums)
```

# Question 4

```python
import numpy as np
```

```python
scores = np.random.randint(50, 101, size=(10, 5))
```

```python
average_scores = np.mean(scores, axis=1)
```

```python
max_scores = np.max(scores, axis=0)
```

```python
final_array = np.column_stack((scores, average_scores))
```

```python
print("Original Scores (10 students × 5 subjects):")
print(scores)
```

```python
print("\nAverage score per student:")
print(average_scores)
```

```python
print("\nMaximum score per subject:")
print(max_scores)
```

```python
print("\nFinal array with average added:")
print(final_array)
```

# Question 6

```python
import pandas as pd

data = {
    "product": ["pen", "pencil", "notebook", "eraser", "marker", "scale", "sharpener", "glue"],
    "Price": [1.5, 0.5, 2.0, 0.75, 1.8, 1.2, 0.9, 1.0],
    "quantity": [10, 25, 15, 30, 12, 18, 20, 22]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['TotalValue'] = df['Price'] * df['quantity']
print(df)
```

```python
df = df.sort_values(by='TotalValue', ascending=False)
print(df)
```

```python
avg_quantity = df.groupby('Price')['quantity'].mean()
print(avg_quantity)
```

# Question 10

```python
def check_bmi(weight, height):
    bmi = weight / (height * height)  # BMI = weight ÷ (height × height)
    if bmi < 18.5:
        return "BMI is " + str(round(bmi, 2)) + " → Underweight"
    elif bmi < 25:
        return "BMI is " + str(round(bmi, 2)) + " → Normal"
    elif bmi < 30:
        return "BMI is " + str(round(bmi, 2)) + " → Overweight"
    else:
        return "BMI is " + str(round(bmi, 2)) + " → Obese"
```

```python
print(check_bmi(60, 1.7))  # 60 kg, 1.7 m tall
print(check_bmi(80, 1.7))  # 80 kg, 1.7 m tall
print(check_bmi(45, 1.6))  # 45 kg, 1.6 m tall
```
